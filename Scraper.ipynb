```python
import requests
from requests import get
from bs4 import BeautifulSoup
from selenium import webdriver
import pandas as pd
import numpy as np
from time import sleep
from random import randint
```

```python
# import time
# start_time = time.time()

urls_collected = []

page = "https://www.tripadvisor.com/Restaurants-g60763-New_York_City_New_York.html"
driver = webdriver.Chrome()
driver.get(page)
sleep(randint(10,15))
driver.find_element_by_link_text('Reserve a Table Online').click()
sleep(randint(7,10))
for i in driver.find_elements_by_class_name('wQjYiB7z'):
    url = i.find_element_by_css_selector('a').get_attribute('href')
    urls_collected.append(url)
print(urls_collected)

print("\n--- %s seconds ---" % (time.time() - start_time))
```

```python
data = {
    'Restaurant' : [],
    'Website' : [],
    'Address' : [],
    'Phone' : [],
    'Email' : [],
    'Reviews' : [],
}
```

```python
import time
start_time = time.time()
for i in range(len(urls_collected)):
    url2 = urls_collected[i]
    driver2 = webdriver.Chrome()
    driver2.get(url2)  
    sleep(randint(5,10))
    try:
        res_name = driver2.find_element_by_class_name('_3a1XQ88S').get_attribute('textContent')
    except:
        driver2.refresh()
        sleep(randint(5,10))
        res_name = driver2.find_element_by_class_name('_3a1XQ88S').get_attribute('textContent')
    try:
        web = driver2.find_element_by_link_text('Website').get_attribute('href')
    except:
        web = 'Not available'
    add = driver2.find_element_by_xpath("//span[@class = '_2saB_OSe']").get_attribute('textContent')
    ph = driver2.find_element_by_xpath("//span[@class = '_15QfMZ2L']").get_attribute('textContent')
    try:
        mail = driver2.find_element_by_link_text('Email').get_attribute('href')
        mail = mail.split(':')[1].split('?')[0]
    except:
        mail = 'Not available'
    rating = driver2.find_element_by_xpath("//a[@class = '_10Iv7dOs']").get_attribute('textContent')

    data['Restaurant'].append(res_name)
    data['Website'].append(web)
    data['Address'].append(add)
    data['Phone'].append(ph)
    data['Email'].append(mail)
    data['Reviews'].append(rating)

    driver2.close()
    
print("--- %s seconds ---" % (time.time() - start_time))
```

```python
df = pd.DataFrame(data)
df.index = np.arange(1, len(df) + 1)
```
